```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# matplotlib
%matplotlib inline
pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60)
pd.set_option('display.mpl_style', 'default')

i_rng = pd.date_range('2015-01-01', '2015-01-01 23:45:00', freq='15min').time
df = pd.read_csv('fulltest.csv', parse_dates=[1,3], dayfirst=True, usecols=[1,3,4,6,7], na_values=0, names=['Duration', 'e_date', 'e_id', 's_date', 's_id'], header=False).dropna()
top5 = df.s_id.value_counts()[:3].index.tolist()

sdays = len(df['s_date'].apply(lambda t: t.date()).unique())
edays = len(df['e_date'].apply(lambda t: t.date()).unique())
i = 14
```

```python
# plt.rcParams['figure.figsize'] = (15, 15)
# total[[26, 27]].plot()
# top5

def sort():
    sw_time = df.s_date.apply(lambda x: x.weekday() < 5) & (df.s_id == i)
    ew_time = df.e_date.apply(lambda x: x.weekday() < 5) & (df.e_id == i)
    s_counts = df.loc[sw_time, ['s_id', 's_date']]
    e_counts = df.loc[ew_time, ['e_date', 'e_id']]
    if(len(s_counts) > 0):
        s_counts.set_index(['s_date'], inplace=True)
        group = s_counts.groupby(pd.TimeGrouper(freq='15Min', nperiods=96)).sum()
        group.set_index(group.index.time, inplace=True)
        icount[i] = group.groupby(group.index).sum()
    if(len(e_counts) > 0):
        e_counts.set_index(['e_date'], inplace=True)
        group2 = s_counts.groupby(pd.TimeGrouper(freq='15Min', nperiods=96)).sum()
        group2.set_index(group2.index.time, inplace=True)
        jcount[i] = group2.groupby(group2.index).sum()
    return icount

def sortB(weektime):
    icount = pd.DataFrame(index=i_rng, columns=range(1, 781))
    jcount = pd.DataFrame(index=i_rng, columns=range(1, 781))
    if weektime != 3:
        if weektime == 0:
            sw_time = df.s_date.apply(lambda x: x.weekday() < 5)
            ew_time = df.e_date.apply(lambda x: x.weekday() < 5)
        if weektime == 1:
            sw_time = df.s_date.apply(lambda x: x.weekday() == 5)
            ew_time = df.e_date.apply(lambda x: x.weekday() == 5)
        if weektime == 2:
            sw_time = df.s_date.apply(lambda x: x.weekday() == 6)
            ew_time = df.e_date.apply(lambda x: x.weekday() == 6)
        s_counts = df.loc[sw_time, ['s_id', 's_date']]
        e_counts = df.loc[ew_time, ['e_date', 'e_id']]
        sdays = len(s_counts['s_date'].apply(lambda t: t.date()).unique())
        edays = len(e_counts['e_date'].apply(lambda t: t.date()).unique())
    else:
        s_counts = df[['s_id', 's_date']]
        e_counts = df[['e_date', 'e_id']]
        sdays = len(s_counts['s_date'].apply(lambda t: t.date()).unique())
        edays = len(e_counts['e_date'].apply(lambda t: t.date()).unique())
    for i in range(1, 781):
        i_start = s_counts.s_id == i
        i_end = e_counts.e_id == i
        s1_counts = s_counts.loc[i_start, ['s_id', 's_date']]
        e1_counts = e_counts.loc[i_end, ['e_date', 'e_id']]
        if(len(s1_counts) > 0):
            s1_counts.set_index(['s_date'], inplace=True)
            group = s1_counts.groupby(pd.TimeGrouper(freq='15Min', nperiods=96)).sum()
            group.set_index(group.index.time, inplace=True)
            icount[i] = group.groupby(group.index).sum()
            icount[i] = icount[i].apply(lambda y: y / i)
        if(len(e1_counts) > 0):
            e1_counts.set_index(['e_date'], inplace=True)
            group2 = e1_counts.groupby(pd.TimeGrouper(freq='15Min', nperiods=96)).sum()
            group2.set_index(group2.index.time, inplace=True)
            jcount[i] = group2.groupby(group2.index).sum()
            jcount[i] = jcount[i].apply(lambda y: y / i)
    icount = icount.dropna(how='all', axis=1).applymap(lambda x: x / sdays).fillna(0)
    jcount = jcount.dropna(how='all', axis=1).applymap(lambda x: x / edays).fillna(0)
    total = pd.concat([icount, jcount], axis=1, keys=['Entries', 'Exits'])
    total = total.swaplevel(0, 1, axis=1).reindex_axis(range(1, 781), axis=1, level=0)
    return total

sortB(0)
# %timeit method = [sortB(0)]
```
