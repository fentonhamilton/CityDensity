```python
import cycledata as cd
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
cd.pd.set_option('display.width', 5000)
cd.pd.set_option('display.max_columns', 60)
cd.pd.set_option('display.mpl_style', 'default')

# Read Tube Station Counts (2012)
WkEntries = pd.read_csv('WkEntries.csv', index_col=0)
WkExits = pd.read_csv('WkExits.csv', index_col=0)
SatEntries = pd.read_csv('SatEntries.csv', index_col=0)
SatExits = pd.read_csv('SatExits.csv', index_col=0)
SunEntries = pd.read_csv('SunEntries.csv', index_col=0)
SunExits = pd.read_csv('SunExits.csv', index_col=0)

# Read Cycle Counts
november = cd.MonthFile('fulltest.csv')
entries, exits , total = november.ExtractData(0)

# Un-comment to extract all components of data
# sat_entries, sat_exits, sat_total = november.ExtractData(1)
# sun_entries, sun_exits, sun_total = november.ExtractData(2)
```

```python
as_entries = entries.plot(figsize=(12,8), legend=False, title='Average Cycle Station Entries', label='entryplot')
as_exits = exits.plot(figsize=(12,8), legend=False, title='Average Cycle Station Exits', label='exitplot')
tube_entries = WkEntries.plot(figsize=(12,8), legend=False, title='Average Tube Station Entries', label='tubeentry')
tube_exits = WkExits.plot(figsize=(12,8), legend=False, title='Average Tube Station Exits', label='tubeexits')
```

```python
# Top 15 Stations entries
entrylist = entries.sum().order(ascending=False).index[:15]
entries[entrylist].plot(figsize=(12,8), title='Average Station Entries: Top 15 Stations')
exitlist = exits.sum().order(ascending=False).index[:15]
exits[exitlist].plot(figsize=(12,8), title='Average Station Exits: Top 15 Stations')
```

```python
SatEntries.plot(figsize=(12,8), title='Saturday Average Station Entries', legend=False)
SatExits.plot(figsize=(12,8), title='Saturday Average Station Exits', legend=False)
SunEntries.plot(figsize=(12,8), title='Sunday Average Station Entries', legend=False)
SunExits.plot(figsize=(12,8), title='Sunday Average Station Exits', legend=False)
```
