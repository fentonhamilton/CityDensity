```python
import cycledata as cd
import matplotlib.pyplot as plt

%matplotlib inline
cd.pd.set_option('display.width', 5000)
cd.pd.set_option('display.max_columns', 60)
cd.pd.set_option('display.mpl_style', 'default')

november = cd.MonthFile('fulltest.csv')
november.df
```

```python
entries, exits ,total = november.ExtractData(0)
sat_entries, sat_exits, sat_total = november.ExtractData(1)
sun_entries, sun_exits, sun_total = november.ExtractData(2)
```

```python
as_entries = entries.plot(figsize=(12,8), legend=False, title='Average Station Entries', label='entryplot')
as_exits = exits.plot(figsize=(12,8), legend=False, title='Average Station Exits', label='exitplot')
entries.legend()
```

```python
# Top 15 Stations entries
entrylist = entries.sum().order(ascending=False).index[:15]
entries[entrylist].plot(figsize=(12,8), title='Average Station Entries: Top 15 Stations')
exitlist = exits.sum().order(ascending=False).index[:15]
exits[exitlist].plot(figsize=(12,8), title='Average Station Exits: Top 15 Stations')
```
